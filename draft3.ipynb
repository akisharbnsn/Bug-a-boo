```python
from pathlib import Path
import pdfplumber
DATA_DIR = Path("../Bug-a-boo/random")
print("PDF files found:", len(list(DATA_DIR.glob("*.pdf"))))

pdf_path = sorted(DATA_DIR.glob("*.pdf"))[0]
with pdfplumber.open(pdf_path) as pdf:
    pages = [(p.extract_text() or "") for p in pdf.pages]
raw = "\n".join(pages)
print("Characters extracted:", len(raw))
print(raw[:800])  # peek sample
if len(raw.strip()) == 0:
    print("Looks like NO text. This PDF is likely scanned images. Switch to OCR (pytesseract + pdf2image).")






```

```python
def extract_text_from_pdf(pdf_path):    
    """Extracts text from a given PDF file."""    
    text = ""    
    with pdfplumber.open(pdf_path) as pdf:        
        for page in pdf.pages:            
            text += page.extract_text() + "\n"    
    return text if text else "No text found in the PDF."

print(extract_text_from_pdf(pdf_path))
```
